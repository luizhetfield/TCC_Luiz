```python
import pandas as pd
import mysql.connector

db = mysql.connector.connect(
    host = "localhost",
    user = "root",
    passwd="mydbwork",
    db = "puc"
)

mycursor = db.cursor()

mycursor.execute('SELECT TXT_CRGO from cadastro')

dados = mycursor.fetchall()

prefix_profissao=[]
df = pd.DataFrame(dados, columns=['TXT_CRGO'])
df["TXT_CRGO"] = df["TXT_CRGO"].astype('string')
```

```python
print("O DF, atualmente, possui %s profissões diversas cadastradas" % str(len(df["TXT_CRGO"].unique())))
```

```python
for i in df.itertuples():
    print("olhando", i.TXT_CRGO, "e procurando por", i.TXT_CRGO[0:4])
    if i.TXT_CRGO[0:4] in prefix_profissao:
        pass
    else:
        print("não está na lista", i.TXT_CRGO)
        if i.TXT_CRGO[0:3] == 'AUX':
            if 'AUX' not in prefix_profissao:
                prefix_profissao.append('AUX')    
        elif i.TXT_CRGO[0:1] in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
            if 'MILITAR' not in prefix_profissao:
                prefix_profissao.append('MILITAR')
        else:
            prefix_profissao.append(i.TXT_CRGO[0:4])
    
```

```python
print(prefix_profissao)
print("A lista prefix_profissao possui %s elementos" % str(len(prefix_profissao)))
print("Já o DF possui %s profissões esparsas" % str(len(df["TXT_CRGO"].unique())))
```

```python
for i in range(len(prefix_profissao)):
    for j in df.itertuples():
        if j.TXT_CRGO is None:
            pass
        elif j.TXT_CRGO == "":
             df.at[j.Index , 'TXT_CRGO'] = "desconhecido"            
        elif j.TXT_CRGO[0:4] == prefix_profissao[i]:
            df.at[j.Index , 'TXT_CRGO'] = prefix_profissao[i]
        elif j.TXT_CRGO[0:3] == 'AUX':
            if j.TXT_CRGO[0:3] == prefix_profissao[i]:
                df.at[j.Index, 'TXT_CRGO'] = prefix_profissao[i]
        else:
            pass
```
